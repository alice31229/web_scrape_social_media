```python
# https://www.linkedin.com/pulse/webscrape-google-map-reviews-using-selenium-python-choy-siew-wearn/

import re
import time
import pandas as pd

from selenium import webdriver
from selenium.webdriver import ActionChains
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options

```

```python
# kwd and get kwd's all chain stores （正 中立 反 -> 1 2 1）

# kwds = ['王品牛排','酒吧','壽喜燒','日式拉麵','鼎泰豐','西提牛排','原燒燒肉','義大利麵','夜市',
#         '觀光景點','摩斯漢堡','河濱公園','超市']

kwds = ['陶板屋','燒肉','肯德基','麥當勞','馬辣','饗食天堂','築間']

# 初始化WebDriver，這裡使用Chrome作為例子
options=Options()
options.add_argument('disable-infobars')
options.add_argument("--disable-blink-features=AutomationControlled")
options.add_argument("--disable-extensions")
options.add_argument("--profile-directory=Default")
options.add_argument("--incognito")
options.add_argument("--disable-plugins-discovery")
options.add_argument("--start-maximized")
options.chrome_executable_path='../webCrawl/chromedriver'
driver=webdriver.Chrome(options=options)

stores = []
urls = []

def get_restaurants_urls(kwds_lst):
    
    for k in kwds_lst:
    
        # 前往目標網頁
        driver.get("https://www.google.com.tw/maps/@23.546162,120.6402133,8z?hl=zh-TW&entry=ttu")

        # 找到搜尋欄的元素，這裡使用XPath作為例子
        search_input = driver.find_element(By.XPATH, "//input[@name='q']")

        # 在搜尋欄輸入內容
        search_input.send_keys(k)

        # 按下Enter鍵進行搜尋
        search_input.send_keys(Keys.ENTER)
        time.sleep(10)

        # 可以繼續處理後續操作，例如等待結果載入、擷取結果等
        restaurants = driver.find_elements(By.XPATH, "//a[@class='hfpxzc']")
        for r in restaurants:
            stores.append(r.get_attribute("aria-label"))
            urls.append(r.get_attribute("href"))
            
            print(r.get_attribute("aria-label"))
            print(r.get_attribute("href"))
        
    # 關閉瀏覽器視窗
    driver.quit()
    
    data = {'stores':stores, 
            'urls':urls}
    stores_comments = pd.DataFrame(data)
    
    return stores_comments

s_and_u = get_restaurants_urls(kwds)
s_and_u
```

```python
# 初始化WebDriver，這裡使用Chrome作為例子
options=Options()
options.add_argument('disable-infobars')
options.add_argument("--disable-blink-features=AutomationControlled")
options.add_argument("--disable-extensions")
options.add_argument("--profile-directory=Default")
options.add_argument("--incognito")
options.add_argument("--disable-plugins-discovery")
options.add_argument("--start-maximized")
options.chrome_executable_path='../webCrawl/chromedriver'
driver=webdriver.Chrome(options=options)

name_list = []
stars_list = []
review_list = []
duration_list = []

review = pd.DataFrame()

def each_store_rating_comment(scroll_cnt):
    
    SCROLL_PAUSE_TIME = 5

    # Get scroll height
    last_height = driver.execute_script("return document.body.scrollHeight")

    number = 0

    while True:
        
        #driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
        number+=1

        # Scroll down to bottom 
        ele = driver.find_element(By.XPATH, '//*[@id="QA0Szd"]/div/div/div[1]/div[2]/div/div[1]/div/div/div[3]')
        driver.execute_script('arguments[0].scrollBy(0, 5000);', ele)

        # Wait to load page

        time.sleep(SCROLL_PAUSE_TIME)

        # Calculate new scroll height and compare with last scroll height
        #print(f'last height: {last_height}')

        ele = driver.find_element(By.XPATH, '//*[@id="QA0Szd"]/div/div/div[1]/div[2]/div/div[1]/div/div/div[3]')
        new_height = driver.execute_script("return arguments[0].scrollHeight", ele)

        #print(f'new height: {new_height}')

        if number == scroll_cnt:
            break

        last_height = new_height
        
    time.sleep(15)
    
    item = driver.find_elements(By.XPATH, '//*[@id="QA0Szd"]/div/div/div[1]/div[2]/div/div[1]/div/div/div[3]')

    time.sleep(3)
    
    for i in item:

        button = i.find_elements(By.TAG_NAME,'button')
        for m in button:
            if m.text == "全文":
                try:
                    m.click()
                except:
                #    pass
                    # 使用ActionChains定位到按鈕上方並點擊
                    actions = ActionChains(driver)
                    actions.move_to_element(m).click().perform()
                    
        time.sleep(10)

        # 可只留評分與評論
        name = i.find_elements(By.CLASS_NAME,"d4r55")
        stars = i.find_elements(By.CLASS_NAME,"kvMYJc")
        #review = i.find_elements(By.CLASS_NAME,"wiI7pd")
        review = i.find_elements(By.XPATH, '//span[@class="wiI7pd"]')
        duration = i.find_elements(By.CLASS_NAME,"rsqaWe")

        for j,k,l,p in zip(name,stars,review,duration):
            name_list.append(j.text)
            stars_list.append(p.text)
            review_list.append(k.get_attribute("aria-label"))
            duration_list.append(l.text)
            
    review = pd.DataFrame(
    {'name': name_list,
     'rating': stars_list,
     'review': review_list,
     'duration': duration_list})
    
    print(review.shape)

    

def get_rating_comments(name_url_df):
    
    for ind in range(len(name_url_df['stores'])):
    
        # 進連結後 按評論按鈕
        driver.get(name_url_df['urls'][ind])
        comment_btn = driver.find_element(By.XPATH, f"//button[@aria-label='對「{name_url_df['stores'][ind]}」的評論']")
        comment_btn.click()
        
        time.sleep(5)
        
        # 執行之前一系列流程
        each_store_rating_comment(20)

get_rating_comments(s_and_u)
        
    
```

```python
review = pd.DataFrame(
    {'name': name_list,
     'rating': stars_list,
     'review': review_list,
     'duration': duration_list})

review
```

```python
review['duration_test'] = review['duration'].str.replace(r'(https?://\S+)|\n', '', regex=True)
pattern = re.compile(r'[^\w\s!@#$%^&*()_+\-=\[\]{};\'\\:"|<,./<>?`~]')
review['duration_final'] = review['duration_test'].apply(lambda x: re.sub(pattern, '', x))

review_final = review.filter(['review','duration_final'])
review_final.columns = ['Rating','Comments']
review_final
```
